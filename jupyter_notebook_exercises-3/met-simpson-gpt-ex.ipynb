# Integral Usando M√©todo de Simpson

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fun√ß√£o a ser integrada
def f(x):
    return np.sin(x) + 2

# Intervalo de integra√ß√£o
a, b = 0, np.pi
n = 6  # O n√∫mero de subintervalos deve ser par

# Divis√£o do intervalo em pontos de amostragem
x = np.linspace(a, b, n+1)
y = f(x)

# Criando pontos intermedi√°rios para melhor visualiza√ß√£o da curva
x_fine = np.linspace(a, b, 100)
y_fine = f(x_fine)
```

```python
# Plotando a fun√ß√£o original
plt.figure(figsize=(8, 5))
plt.plot(x_fine, y_fine, 'b-', label=r'$f(x) = \sin(x) + 2$')

# Aplicando o M√©todo de Simpson para aproximar a integral
for i in range(0, n, 2):
    x_parabola = np.linspace(x[i], x[i+2], 20)
    y_parabola = f(x[i]) * (x_parabola - x[i+1]) * (x_parabola - x[i+2]) / ((x[i] - x[i+1]) * (x[i] - x[i+2])) + \
                 f(x[i+1]) * (x_parabola - x[i]) * (x_parabola - x[i+2]) / ((x[i+1] - x[i]) * (x[i+1] - x[i+2])) + \
                 f(x[i+2]) * (x_parabola - x[i]) * (x_parabola - x[i+1]) / ((x[i+2] - x[i]) * (x[i+2] - x[i+1]))
    
    plt.fill_between(x_parabola, y_parabola, alpha=0.3, color='orange')

# Marcando os pontos usados no m√©todo de Simpson
plt.scatter(x, y, color='red', zorder=3, label="Pontos de amostragem")

# Ajustes do gr√°fico
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Aproxima√ß√£o da Integral pelo M√©todo de Simpson")
plt.legend()
plt.grid()

# Exibir o gr√°fico
plt.show()
```
