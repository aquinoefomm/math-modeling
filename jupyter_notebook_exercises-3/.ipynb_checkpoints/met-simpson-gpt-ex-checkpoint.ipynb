# Integral Usando Método de Simpson

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Função a ser integrada
def f(x):
    return np.sin(x) + 2

# Intervalo de integração
a, b = 0, np.pi
n = 6  # O número de subintervalos deve ser par

# Divisão do intervalo em pontos de amostragem
x = np.linspace(a, b, n+1)
y = f(x)

# Criando pontos intermediários para melhor visualização da curva
x_fine = np.linspace(a, b, 100)
y_fine = f(x_fine)
```

```python
# Plotando a função original
plt.figure(figsize=(8, 5))
plt.plot(x_fine, y_fine, 'b-', label=r'$f(x) = \sin(x) + 2$')

# Aplicando o Método de Simpson para aproximar a integral
for i in range(0, n, 2):
    x_parabola = np.linspace(x[i], x[i+2], 20)
    y_parabola = f(x[i]) * (x_parabola - x[i+1]) * (x_parabola - x[i+2]) / ((x[i] - x[i+1]) * (x[i] - x[i+2])) + \
                 f(x[i+1]) * (x_parabola - x[i]) * (x_parabola - x[i+2]) / ((x[i+1] - x[i]) * (x[i+1] - x[i+2])) + \
                 f(x[i+2]) * (x_parabola - x[i]) * (x_parabola - x[i+1]) / ((x[i+2] - x[i]) * (x[i+2] - x[i+1]))
    
    plt.fill_between(x_parabola, y_parabola, alpha=0.3, color='orange')

# Marcando os pontos usados no método de Simpson
plt.scatter(x, y, color='red', zorder=3, label="Pontos de amostragem")

# Ajustes do gráfico
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Aproximação da Integral pelo Método de Simpson")
plt.legend()
plt.grid()

# Exibir o gráfico
plt.show()
```
