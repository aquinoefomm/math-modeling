# MÃ©todo de Euler - exemplo da video aula

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# EDO:
f = lambda t, s: np.exp(-t)

# intervalo:
h = 0.1

# tempo discretizado:
t = np.arange(0, 1 + h, h)

# condiÃ§Ã£o inicial:
s0 = -1

#mÃ©todo explÃ­cito de Euler:
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

#grÃ¡fico:
plt.figure(figsize = (12, 8))
plt.plot(t, s, "r--", label="AproximaÃ§Ã£o")
plt.plot(t, -np.exp(-t), "g", label="Exato")
plt.title("SoluÃ§Ã£o Exata e Aproximada de EDO")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.grid()
plt.legend(loc="lower right")
plt.show()
```
