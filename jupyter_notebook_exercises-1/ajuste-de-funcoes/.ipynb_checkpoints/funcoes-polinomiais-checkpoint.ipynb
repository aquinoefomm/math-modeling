# Ajuste de funções

## Ajuste de curva

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.polynomial.polynomial as poly
```

```python
# Solução de sistema linear pelo método de Gauss:
def SubRet(U, c):
    n = c.size
    x = np.zeros(n)
    for i in reversed(range(n)):
        x[i] = (c[i] -U[i,i+1:]@x[i+1:])/U[i,i]
    return x

def ElimGauss(A, b):
    U = np.copy(A)
    c = np.copy(b)
    n = c.size
    for j in range(n-1):
        for i in range(j+1, n):
            m = U[i,j]/U[j,j]
            U[i,j:] -= m*U[j,j:]
            c[i] -= m*c[j]
    x = SubRet(U,c)
    return x
```

```python
# Dados do problema:
x = np.arange(1.0, 8.0)
y = np.array([1., 3., 2., 4., 2., 5., 4.])
```

```python
def Ajuste_Linear(x,y):
    # definir matriz A:
    m = x.size
    A = np.ones((m,2))
    for k in reversed(range(m)):
        A[k,0] = x[k]
    x = ElimGauss(A.T@A, A.T@y)
    return x
```

```python
# Solução:
# Obter os coeficientes da função y = b + ax
p = Ajuste_Linear(x,y)
plt.plot(x,y,"o")
# Obtendo os valores da reta ajustada:
x_novo = np.arange(0.0, 9.1, 0.1)
plt.plot(x_novo, np.polyval(p, x_novo))
plt.title("Ajuste Linear")
plt.tight_layout()
plt.show()
```

## Ajuste de funções polinomiais:


```python
def ajuste_poli(x, y, n):
    m = x.size
    A = np.ones((m , n+1))
    for k in reversed(range(n)):
        A[:,k] = x**(n - k)
    p = ElimGauss(A.T@A, A.T@y)
    return p

```

```python
for i in range(1, 5):
    p = ajuste_poli(x, y, 2*i - 1)
    plt.subplot(2, 2, i)
    plt.plot(x, y, "o")
    x_novo = np.arange(0.0, 8.1, 0.1)

    plt.plot(x_novo, np.polyval(p, x_novo))
    plt.title(f'Ordem do Polinômio {2*i - 1}')
    plt.tight_layout()
    plt.show()
```
