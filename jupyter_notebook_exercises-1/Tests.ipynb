```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 2*x**3 + x**2 + 200
```

```python
f(1)
```

```python
lista_x = np.linspace(-10, 10, 100)
lista_y = f(lista_x)
plt.grid(True)
plt.plot(lista_x, lista_y)
```
