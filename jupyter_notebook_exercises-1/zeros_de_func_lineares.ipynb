```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 2*x**3 - 11.7*x**2 + 17.7*x - 5
```

```python
x_anterior = 0
x_depois = 4
```

```python
lista_x = np.linspace(x_anterior, x_depois, 1001)
lista_y = f(lista_x)
```

```python
plt.style.use("dark_background")
plt.plot(lista_x, lista_y)
plt.grid(True)
```

```python
def df(x):
    return 6*x**2 - 23.4*x + 17.7
```

```python
def newton(chute, iteracoes):
    raiz = chute
    for i in range(iteracoes):
        raiz = raiz - f(raiz)/df(raiz)
    return raiz

print(newton(3, 3))
```

```python
def secante(a, b, iteracoes):
    x_0 = a
    x_1 = b
    for i in range(iteracoes):
        chute = x_0 - f(x_0) * (x_1 - x_0) / (f(x_1) - f(x_0))
        x_0 = x_1
        x_1 = chute
    erro_rel = (x_1 - x_0) / x_1 * 100
    return x_1, '{:.2f}%'.format(erro_rel)

print(secante(3, 4, 3))
```
